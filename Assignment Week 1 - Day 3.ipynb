*   multicollinearity is when two independent variables in the data are highly correlated with each other means there is relationship between both variables it can be positive or negative.
multicollinearity is against the basic assumptions of linear regression algorithm.

*   There are various ways of treating multicollinearity:-
1)removing all highly correlated variables ( we can find out highly correlated feature by VIF i.e. higher the value of VIF for the feature more linearly correlated that feature is.
2)we can linearly combine independent variables by adding them together.
3)use of principal component analysis to create a set of uncorrelated variables to include in model.

## 2. VIF(Variation Inflation Factor),Weight of Evidence & Information Value. Why and when to use?

*   VIF is simply variation inflation factor and it measures the amount of multicollinearity in the regression analysis.

*   Weight of Evidence:- The weight of evidence tells us the predictive power of an independent variable in relation to the dependent variable. Since it evolved from credit scoring world, It generally describes as measure of the separation of "Good customers" and "Bad customers".

 "Good customers" refers to customers who paid back loan. and "Bad customers" refers to customers who defaulted the loan.
 `WOE=ln(Distribution of Goods/Distribution of Bads)`

*   positive WOE means Distribution of goods>distribution of bads.
negative WOE means Distribution of goods < distributions of bads.

 WOE=ln( % of non-events / % of events)

*  IV (Information Value) is one of the most useful technique to select important variables in a predictive model. It helps to rank variables on the basis of their importance.
  
  A high IV value indicates that the independent variable is a strong predictor  of the dependent variable, while a low IV value indicates that the independent variable does not contribute much to the predictive power of the model.
  The IV is calculated using the following formula:
 `IV= (% of non-events - % of events) * WOE`



## 3. What is tradeoff between bias and variances, and the relationship between them?

*  The gap between the model - predicted values and the actual values is referred to as bias. It's a mistake. A low bias is one of the objectives of an ML algorithm. *  The vulnerability of the model to tiny changes in the training dataset is referred to as variance. Low variance is another goal of an ML algorithm. 
*  It is impossible to have both low bias and low variance in a dataset that is not perfectly linear. The variance of a straight line model is low, but the bias is large, whereas the variance of a high-degree polynomial is low, but the bias is high. In machine learning, the link between bias and variation is unavoidable.

* Decreasing the bias increases the variance.
* Decreasing the variance increases the bias.
* So, there is a trade-off between the two; the ML specialist has to decide, based on the assigned problem, how much bias and variance can be tolerated. Based on this, the final model is built.



## 4. How overfitting and underfitting can affect model generalization?



* Overfitting and underfitting are two common problems that can affect a machine learning model's ability to generalize to new, unseen data.

* Overfitting occurs when a model is too complex and learns the noise in the training data, instead of the underlying patterns. As a result, the model performs well on the training data, but poorly on the  test data, because it has essentially memorized the training data and cannot generalize to new data. Overfitting can be optimised by reducing the complexity of the model or by adding regularization techniques such as dropout, weight decay, or early stopping.

* Underfitting occurs when a model is too simple and cannot capture the underlying patterns in the training data. As a result, the model performs poorly on both the training and validation or test data. Underfitting can be optimised by increasing the complexity of the model, using a different model architecture, or by increasing the size of the training data.

* both overfitting and underfitting can have a negative impact on model generalization. Overfitting leads to poor performance on new data because the model has memorized the training data and cannot generalize to new data, and underfitting leads to poor performance because the model is too simple and cannot capture the underlying patterns in the data. Therefore, it is important to find a balance between model complexity and performance on the validation or test data to ensure good generalization performance.
